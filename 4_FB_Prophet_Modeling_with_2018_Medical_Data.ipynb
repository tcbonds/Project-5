```python
import numpy as np
import pickle
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.metrics import mean_squared_error
from fbprophet import Prophet
```

```python
# Pickling Call Volume all df
redo_pickle = False

if redo_pickle == True:
    with open('hourly_call_volume_medical_2018.pickle','wb') as f:
        pickle.dump(hourly_call_volume_medical_2018,f)
        
if redo_pickle == False:
    with open('hourly_call_volume_medical_2018.pickle','rb') as f:
        hourly_call_volume_medical_2018 = pickle.load(f)
```

```python
# Preprocessing data for prophet
hourly_call_volume_medical_2018 = pd.DataFrame(hourly_call_volume_medical_2018).reset_index()\
.rename(columns={'Hourly_DtTm':'ds','Call Number':'y'})
```

```python
# Creating 80-20 train-test split
size = int(len(hourly_call_volume_medical_2018) * 0.8)
train, test = hourly_call_volume_medical_2018[0:size], hourly_call_volume_medical_2018[size:]
```

```python
# Instantiating Prophet Model
m = Prophet(daily_seasonality=True, weekly_seasonality=True)
# Training Model
m.fit(train)
```

```python
# Creating Dataframe for Predicting 2 Weeks into Future
future = m.make_future_dataframe(periods=14)
future.tail()
```

```python
# Predicting 2 Weeks into Future
forecast = m.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
# Plotting 2 week prediction into future
fig1 = m.plot(forecast)
```

```python
# Finding RMSE of prediction vs actual
np.sqrt(mean_squared_error(test.iloc[:14,1], forecast.loc[7004:,'yhat']))
# RMSE is higher than previously made models possibly 
# because Prophet is such a conservative model
```

```python
# Plotting Seasonality Components
fig2 = m.plot_components(forecast)
```

```python
# Plotting the Number of Calls on Each Day over 2018
hourly_call_volume_medical_2018['dow'] = hourly_call_volume_medical_2018.ds.dt.dayofweek
sns.lineplot(data=hourly_call_volume_medical_2018.groupby('dow').sum())
plt.xticks(range(7), ['M','Tu','W','Th','F','Sa','Su']);

# This directly contradicts Prophet's seasonality components plots above
# which indicates that Prophet is not working well with my data
```
