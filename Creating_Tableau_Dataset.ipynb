```python
weather = pd.read_csv('sf_weather_98_to_present.csv')
```

```python
# Adding columns for hour of day and day of week

amb_df['hour'] = amb_df['Received DtTm'].dt.hour
amb_df['day_of_week'] = amb_df['Received DtTm'].dt.dayofweek
```

```python
# Creating dictionary to map days of week onto numerical values
day_of_week_dict={0:'Monday',1:'Tuesday', 2:'Wednesday',3:'Thursday',4:'Friday',5:'Saturday',6:'Sunday'}

# Mapping dicitonary values onto dataframe
amb_df['day_of_week'] = amb_df['day_of_week'].map(day_of_week_dict)
```

```python
# Adding column with combined datetime values which 
# will correspond with Hourly_DtTm column in amb_weather_df
weather['date_time'] = pd.to_datetime(weather.Date) + weather.Hour.astype('timedelta64[h]')
```

```python
# Merging dataframes
amb_weather_df = amb_df.merge(weather, left_on='Hourly_DtTm', right_on='date_time')
```

```python
# Saving as csv
amb_weather_df.to_csv('amb_weather.csv')
```
